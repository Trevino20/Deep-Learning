# Batch - Normalization

Batch-Normalization (BN) is an algorithmic method which makes the training  of deep neural networks (DNN) faster and more stable.
    
It consists of normalizing activation vectors from hidden layer using the mean and variance of the current batch. this normalization step is applied right before (or right after) the nonlinear function.

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
from tensorflow.keras.layers import BatchNormalization
```

```python
model = Sequential()

model.add(Dense(3,activation='relu',input_dim=2))
model.add(BatchNormalization())
model.add(Dense(2,activation='relu'))
model.add(BatchNormalization())
model.add(Dense(1,activation='sigmoid'))

model.summary()
```
