```python
import pandas as pd
import numpy as np
import tensorflow
import keras
from keras.layers import Dense
from keras.models import Sequential
from keras.layers import Dropout
from keras.optimizers import Adam
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
X, y = make_moons(100, noise=0.35,random_state=2)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

# Xavier/Glorot Initialization

```python
from keras.initializers import glorot_normal, glorot_uniform
```

```python
# Example model with Xavier/Glorot Normal Initialization
model = Sequential()
model.add(Dense(10, input_dim=2, kernel_initializer=glorot_normal(), activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.get_weights()
```

```python
# Example model with Xavier/Glorot Uniform Initialization
model_uniform = Sequential()
model_uniform.add(Dense(10, input_dim=2, kernel_initializer=glorot_uniform(), activation='relu'))
model_uniform.add(Dense(1,activation='sigmoid'))
```

```python
model = Sequential()

model.add(Dense(10,activation='relu',input_dim=2,kernel_initializer='he_normal'))
model.add(Dense(10,activation='relu',kernel_initializer='he_normal'))
model.add(Dense(10,activation='relu',kernel_initializer='he_normal'))
model.add(Dense(10,activation='relu',kernel_initializer='he_normal'))
model.add(Dense(1,activation='sigmoid'))

model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history = model.fit(X,y,epochs=100,validation_split=0.2,verbose=0)
```

```python
model.get_weights()
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X,y.astype('int'), clf=model, legend=2)
```
